####  (i) 25% on the orders having final price (discounted price) greater than 600

## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Q5. Find the Total Revenue generated by the E-Commerce company over all orders placed. 

```python
# Calculate profit according to given conditions
def profit(discounted_price):
    if discounted_price > 600:
        profit = discounted_price*0.25
    elif discounted_price > 350:
        profit = discounted_price*0.15
    elif discounted_price > 100:
        profit = discounted_price*0.10
    else:
        profit = discounted_price*0.25
    return profit
```

```python
# Calculate Total_Revenue according to given conditions(We can add overall discounted_price over the profit function)
Total_Revenue = ECOM_df['discounted_price'].apply(profit).sum()
print("Total Revenue generated by the E-Commerce company over all orders placed:", Total_Revenue)
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)

### Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order

```python
# Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order
# 1. We create a new column('BrandRevenue') in ECOM_df dataframe and subtract the discounted_price.....
# ....(which is genrated by the price function) from actual discounted_price.

ECOM_df['BrandRevenue'] = ECOM_df['discounted_price'] - ECOM_df['discounted_price'].apply(profit)
ECOM_df.head(5)
```

* **BrandRevenue** is added to ECOM_df dataframe

```python
# Now we apply group method for  'Brand' and 'BrandRevenue' and calculate each 'BrandRevenue' using sum()method with head(10)
BrandRevenue = ECOM_df.groupby('Brand')['BrandRevenue'].sum().sort_values(ascending=False).head(10)
print("The top 10 Brand having maximum revenue are:")
print(BrandRevenue)
```

## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and 
## iv.	total BrandRevenue generated per Brand. 

## Also, draw a pairplot using these four features. (6 marks)

### Q 8. i. total number of orders placed per Brand
### ii. total retail_price per Brand
### iii. total discounted_price per Brand, and
### iv. total BrandRevenue generated per Brand.

```python
ECOM_df.head()
```

```python
# Create New dataframe and copy Brand column in it using copy() method
NEW_ECOM_df = ECOM_df[['Brand']].copy()
NEW_ECOM_df.head()
```

```python
# Total number of orders placed per Brand in new dataframe using old dataframe
Per_Brand_orders = ECOM_df.groupby('Brand')['Order_Date'].nunique().reset_index()
Per_Brand_orders.rename(columns={'Order_Date':'Orders'},inplace=True)
NEW_ECOM_df = Per_Brand_orders[['Brand','Orders']]
NEW_ECOM_df.head()
```

```python
# Total retail_price per Brand
TotalRetail_Price = ECOM_df.groupby('Brand')['retail_price'].sum().reset_index()

# Merge the 'TotalRetail_Price' column
NEW_ECOM_df = pd.merge(NEW_ECOM_df, TotalRetail_Price, on='Brand')

# Rename the column to reflect the total retail price
NEW_ECOM_df .rename(columns={'retail_price': 'TotalRetail_Price'}, inplace=True)
NEW_ECOM_df.head()
```

```python
# Total discounted_price per Brand
TotalDiscounted_Price = ECOM_df.groupby('Brand')['discounted_price'].sum().reset_index()

# Merge the 'TotalDiscounted_Price' column
NEW_ECOM_df = pd.merge(NEW_ECOM_df, TotalDiscounted_Price, on='Brand')

# Rename the column to reflect the TotalDiscounted_Price
NEW_ECOM_df .rename(columns={'discounted_price': 'TotalDiscounted_Price'}, inplace=True)
NEW_ECOM_df.head()
```

```python
# TotalBrandRevenue generated per Brand.
TotalBrandRevenue = ECOM_df.groupby('Brand')['BrandRevenue'].sum().reset_index()

# Merge the 'TotalBrandRevenue' column
NEW_ECOM_df = pd.merge(NEW_ECOM_df, TotalBrandRevenue, on='Brand')

# Rename the column to reflect the TotalBrandRevenue
NEW_ECOM_df .rename(columns={'BrandRevenue': 'TotalBrandRevenue'}, inplace=True)
NEW_ECOM_df.head()
```

```python
# Draw a pairplot using these four features.
sns.pairplot(data=NEW_ECOM_df)
```

* If a company applies a discount on products, then more products are bought by users, and the company's total brand revenue is positively increased.

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Q9. a) Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.

```python
# Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.
```

```python
ECOM_df.head()
```

```python
# Convert the "order_date" column to a datetime format
ECOM_df['Order_Date'] = pd.to_datetime(ECOM_df['Order_Date'])

# Extract the month and year parts, then combine them to create a new column.
ECOM_df['Month_Year'] = ECOM_df['Order_Date'].dt.strftime('%B-%Y')
ECOM_df.head(5)
```

* Month_Year Column is add into ECOM_df

```python
# Group the DataFrame by Region and Month_Year and sum the Revenue
grouped = ECOM_df.groupby(['Region', 'Month_Year'])['BrandRevenue'].sum().reset_index()

# Iterate over the unique regions and plot the monthly revenue
for region in grouped['Region'].unique():
    region_data = grouped[grouped['Region'] == region]
    
    plt.plot(region_data['Month_Year'], region_data['BrandRevenue'], label=region)
    plt.title(f"Monthly Revenue for {region}")
    plt.xlabel('Month-Year')
    plt.ylabel('BrandRevenue')
    plt.xticks(rotation=90)
    plt.legend()
    plt.show()
```
